```python
import pandas as pd
import numpy as np
from scipy.spatial.distance import cosine
```

```python
df = pd.read_csv("Data06-Oct.csv")
df
```

```python
#ตัด column O ออก เพราะจะทำ user-based collaborative filtering
user = df.drop('O',1)
```

```python
user
```

```python
data_ibs = pd.DataFrame(index=user.columns,columns=user.columns)
data_ibs
```

## Cosine of vector

\begin{equation}
cos(V_A, V_B) = \frac{\sum_{}{A_i, B_i}}{\sqrt{\sum_{}{A_i ^ 2}}\sqrt{\sum_{}{B_i ^ 2}}} \\
When \; A_i \text{is i-th member of}\; V_A\; and \; B_i \text{is i-th member of}\; V_B
\end{equation}

```python
#คำนวนหาค่า cosine similarity
for i in range (0, len(data_ibs.columns)):
    for j in range (0, len(data_ibs.columns)):
        data_ibs.iloc[i,j] = 1-cosine(user.iloc[:,i],user.iloc[:,j])
```
