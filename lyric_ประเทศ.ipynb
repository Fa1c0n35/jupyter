## แก้ตัดคำ ให้ถูก

```
wordlist = pythainlp.corpus.thaiword.get_data()
len(wordlist)
```

```
wordlist.remove('กลางกรุง')
```

```
song = song.replace('สถุน', 'สถุล')
```

```
wordlist.append('แม่ง')
```

```
wordlist.remove('ไม่ผิด')
wordlist.remove('อ่านหนังสือ')
wordlist.remove('จับมือใครดมไม่ได้')
wordlist.remove('ทั้งสอง')
```

```
'บอ' in wordlist
wordlist.remove('บอ')
```

```
trie = create_custom_dict_trie(wordlist)
```

```
words = dict_word_tokenize(song, trie)
len(words)
```

```
words
```

```
cnt = Counter(words)
cnt.most_common(10)
```

```
# เอาพวก space ออก
del cnt['\r\n']
del cnt[' ']
del cnt['\n']
```

```
for k in list(cnt.keys()):
  if len(k)==1:
    del cnt[k]
```

```
df = pd.DataFrame.from_records(cnt.most_common(100), columns=['word', 'freq'])
df_freq = df
df.head(20)
```

## TNC & tfidf

```
!wget -nv http://www.arts.chula.ac.th/ling/wp-content/uploads/tncwordfreq-201712.xlsx
df_tnc = pd.read_excel('tncwordfreq-201712.xlsx', usecols=[0,1], dtype={'Word':str})
```

```
df_tnc = df_tnc.set_index('Word')
df_tnc.head()
```

```
df = df.join(df_tnc, on='word')
```

```
N = df.Freq.max()
```

```
df['tfidf'] = (df.freq * np.log(N/df.Freq)).round(1)
```

```
df = df.sort_values(by='tfidf', ascending=False)
df.head(10)
```
