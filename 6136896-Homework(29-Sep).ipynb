```python
import pandas as pd
from scipy.spatial.distance import cosine
from sklearn.cluster import KMeans
```

```python
df = pd.read_csv("Data29-Sep.csv")
```

```python
df
```

```python
#ตัด column O ออก เพราะเราจะทำ item-based collaborative filtering
items = df.drop('O',1)
```

```python
items
```

```python
data_ibs = pd.DataFrame(index=items.columns,columns=items.columns)
```

```python
data_ibs
```

```python
#คำนวนหาค่า cosine similarityห
for i in range (0, len(data_ibs.columns)):
    for j in range (0, len(data_ibs.columns)):
        data_ibs.iloc[i,j] = 1-cosine(items.iloc[:,i],items.iloc[:,j])
```
