```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
zeta = np.linspace(0, 2*np.pi, 1024)

a,b = 1,1
r = b + a*np.cos(zeta)
x = r*np.cos(zeta)
y = r*np.sin(zeta)

plt.figure(figsize = (5,5))
plt.plot(x,y)

plt.xlim(x.min()*1.1, x.max()*1.1)
plt.ylim(y.min()*1.1, y.max()*1.1)
plt.grid()
plt.show()
```
