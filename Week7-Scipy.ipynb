```python
import numpy as np
import matplotlib.pyplot as plt

def f(x):
    return x*x + 10 * np.sin(x)

x = np.linspace(-10,10,100)

plt.plot(x,f(x))
plt.show()
```

```python
from scipy import optimize

optimize.minimize(f, x0 = 8,method="Nelder-Mead")
```

```python
x = np.linspace(-5,5,100)
y = np.sin(t)  + 0.2* np.random.normal(size = 100)

plt.figure(figsize=(7,7))
plt.scatter(x,y)
```

```python
from scipy import optimize
def f(x, a,b):
    return a*np.sin(b*x)

params, params_covar = optimize.curve_fit(f, x ,y, p0 = [1.1,1.1])
```

```python
print(params)
```

```python
plt.figure(figsize = (7,7))
plt.scatter(x,y, label = 'Original')
plt.plot(x,f(x, params[0], params[1]), label ='corve_fit')
plt.legend(loc = 'best')
plt.show()
```

```python
from scipy import fftpack

x = np.arange(0,20,.02)
y = np.sin(0.4*np.pi*x) + 0.1*np.random.randn(x.size)


#plt.figure(figsize = (10,.6))
plt.plot(x,y)
plt.show()
```

```python
 y_fft = fftpack.fft(y)
power = np.abs(y_fft)
freq = fftpack.fftfreq(y.size, .02)
plt.plot(freq, power)
```

```python
mask = np.where(freq> 0)
freq2 = freq[mask]
peak_freq = freq2[power[mask].argmax()]

print (freq2)
print (peak_freq)

high_freq_fft  =  y_fft.copy()
high_freq_fft[np.abs(freq) >  peak_freq] = 0 
filter_y = fftpack.ifft(high_freq_fft)

plt.figure(figsize = (10,6))
plt.plot(x, filter_y) 
plt.show()
```
