```python
import numpy as np
import matplotlib.pyplot as plt

x = np.linspace(-np.pi, np.pi, 256, endpoint=True)
c, s = np.cos(x), np.sin(x)

plt.plot(x, c, color = 'black', label = 'xc', linewidth = 1.5, linestyle = ':')
plt.plot(x, S, color = 'red', label = 'xs', linewidth = 1.5, linestyle = '-.')
plt.legend(loc = 'best')

plt.xlim(x.min(), x.max()) #-
plt.ylim(c.min(), c.max()) #-

plt.xticks(np.linspace(-np.pi, np.pi, 5, endpoint=True)) # 10 
plt.yticks(np.linspace(-1,1,5, endpoint=True)) # 10 

ax = plt.gca()
ax.spines['right'].set_color('none')
ax.spines['top'].set_color('none')
ax.spines['bottom'].set_position(('data', 0)) # 0
ax.spines['left'].set_position(('data', 0)) # 0

plt.show()
```
