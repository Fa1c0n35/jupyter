```python
import numpy as np
```

```python
data_input = open('2_in.csv','r')
data = np.loadtxt(data_input,delimiter=",",dtype='int',skiprows = 1)

line = data[:,0]
data = (np.c_[data,np.zeros(len(line))]) 

for i in range(0,len(line)):
	data[i][3] = np.divide(data[i][2],data[i][1])

data_process = data[data[:,3].argsort()[::-1]]
data_sort = data_process[:,0].astype('int')
print(data_sort)
```
